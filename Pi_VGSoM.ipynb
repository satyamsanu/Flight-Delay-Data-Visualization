```python
#Delayed
#Delayed if (actual arrival time – scheduled arrival time) / (Scheduled flying time) > 10%. 
df['DELAYED']=(df['ARR_DELAY_NEW']/df['CRS_ELAPSED_TIME'])*100
df['DELAYED'] = np.where((df['DELAYED'] > 10), 1, 0)

#Landing Delayed
#Delayed if (actual arrival time – actual departure time – scheduled flying time) / (scheduled flying time) > 10%.
df['LANDING_DELAYED']=((df['ACTUAL_ELAPSED_TIME']-df['CRS_ELAPSED_TIME'])*100)/df['CRS_ELAPSED_TIME']
df['LANDING_DELAYED'] = np.where((df['LANDING_DELAYED'] > 10), 1, 0)

#Arrival Delay
#Delayed if (actual departure time – scheduled departure time) / (scheduled flying time) > 10%.
df['TAKEOFF_DELAYED']=(df['DEP_DELAY_NEW']/df['CRS_ELAPSED_TIME'])*100
df['TAKEOFF_DELAYED'] = np.where((df['TAKEOFF_DELAYED'] > 10), 1, 0)
```

# Airport Delay

```python
#Airports with Atleast 10 takeoffs and landing per day

df_p=df.copy()
table1=pd.pivot_table(df_p,values='FLIGHTS', index='FL_DATE', columns='ORIGIN', aggfunc=np.sum)
table2=pd.pivot_table(df_p,values='FLIGHTS', index='FL_DATE', columns='DEST', aggfunc=np.sum)
table1=table1.fillna(0)
table2=table2.fillna(0)


#Airports with Atleast 10 takeoffs per day
table11=table1.mean()
table11=table11.to_frame()
table11.columns=['Avg Number of takeoffs per day']

#Airports with Atleast 10 landing per day
table22=table2.mean()
table22=table22.to_frame()
table22.columns=['Avg Number of landing per day']

table=table22.copy()
table['Avg Number of takeoffs per day']=table11['Avg Number of takeoffs per day']
table=table[table['Avg Number of takeoffs per day']>10]
table=table[table['Avg Number of landing per day']>10]

Airports=table.index.tolist()

#Takeoff Delay
df_dp_delay = df.groupby(['ORIGIN'])['DELAYED'].mean()
df_dp_delay=df_dp_delay.to_frame()

#Landing Delay
df_ar_delay = df.groupby(['DEST'])['DELAYED'].mean()
df_ar_delay=df_ar_delay.to_frame()


df_s_delay = pd.concat([df_ar_delay,df_dp_delay], axis=1)
df_s_delay.columns=['Delayed_ar','Delayed_dp']
df_s_delay['Avg_Delay']=(df_s_delay['Delayed_ar']+df_s_delay['Delayed_dp'])/2

#Filter (Only major airports)
df_s_delay_major=df_s_delay.loc[Airports]
```

# Seasonal Delay

```python
#Systematic Delay

#Methods to find Systematic Delay
def Check_Systematic_Delay_Origin(Date,Origin):
    df1=df[df['FL_DATE']==Date]
    df1=df1[df1['ORIGIN']==Origin]
    df1 = df1['DELAYED'].mean()
    if(df1>(1.15*df_s_delay.loc[Origin,'Delayed_dp'])):
        s=1
    else:
        s=0
    return s

def Check_Systematic_Delay_Dest(Date,Dest):
    df1=df[df['FL_DATE']==Date]
    df1=df1[df1['DEST']==Dest]
    df1 = df1['DELAYED'].mean()
    if(df1>(1.15*df_s_delay.loc[Dest,'Delayed_ar'])):
        s=1
    else:
        s=0
    return s

def Check_Systematic_Delay(Date,Origin,Dest):
    s1=Check_Systematic_Delay_Origin(Date,Origin)
    s2=Check_Systematic_Delay_Dest(Date,Dest)
    return (s1 or s2)

def Find_Systematic_Delay_from_Table(Date,Origin,Dest):
    s1=df_qt.loc[Date,Origin]
    s2=df_qt.loc[Date,Dest]
    return (s1 or s2)
```

```python
#Create Systematic Delay Table

#Only major airports
df_major = df[df.ORIGIN.isin(Airports)]
df_major = df_major[df_major.DEST.isin(Airports)]

df_qt = df_major.groupby(['FL_DATE'])['DELAYED'].mean()
df_qt=df_qt.to_frame()
df_qt = df_qt.drop("DELAYED",1)
df_qt=pd.concat([df_qt,pd.DataFrame(columns=df_s_delay_major.index.tolist())])


columns = df_qt.columns.tolist()
for i in columns:
    for index, row in df_qt.iterrows():
        df_qt.loc[index,i]=Check_Systematic_Delay(index,i,i)
```

```python
#Delay Distribution, estimating the Airline Ranking
from scipy.optimize import curve_fit

#___________________________________________
# Model function used to fit the histograms
def func(x, a, b):
    return a * np.exp(-x/b)
#-------------------------------------------
points = [] ; label_company = []
fig = plt.figure(1, figsize=(11,11))
i = 0
for carrier_name in data['UNIQUE_CARRIER'].unique():
    i += 1
    ax = fig.add_subplot(5,3,i)    
    #_________________________
    # Fit of the distribution
    n, bins, patches = plt.hist(x = data[data['UNIQUE_CARRIER']==carrier_name]['DEP_DELAY'],
                                range = (15,180), bins= 60)
    bin_centers = bins[:-1] + 0.5 * (bins[1:] - bins[:-1])    
    popt, pcov = curve_fit(func, bin_centers, n, p0 = [1, 2])
    #___________________________
    # bookeeping of the results
    points.append(popt)
    label_company.append(carrier_name)
    #______________________
    # draw the fit curve
    plt.plot(bin_centers, func(bin_centers, *popt), 'r-', linewidth=3)    
    #_____________________________________
    # define tick labels for each subplot
    if i < 10:
        ax.set_xticklabels(['' for x in ax.get_xticks()])
    else:
        ax.set_xticklabels(['{:2.0f}h{:2.0f}m'.format(*[int(y) for y in divmod(x,60)])
                            for x in ax.get_xticks()])
    #______________
    # subplot title
    plt.title(carrier_name, fontsize = 14, fontweight = 'bold', color = 'darkblue')
    #____________
    # axes labels 
    if i == 4:
        ax.text(-0.3,0.9,'Normalized count of flights', fontsize=16, rotation=90,
            color='k', horizontalalignment='center', transform = ax.transAxes)
    if i == 14:
        ax.text( 0.5, -0.5 ,'Delay at origin', fontsize=16, rotation=0,
            color='k', horizontalalignment='center', transform = ax.transAxes)
    #___________________________________________
    # Legend: values of the a and b coefficients
    ax.text(0.68, 0.7, 'a = {}\nb = {}'.format(round(popt[0],2), round(popt[1],1)),
            style='italic', transform=ax.transAxes, fontsize = 9.5, family='fantasy',
            bbox={'facecolor':'tomato', 'alpha':0, 'pad':5})
    
plt.tight_layout()
```

This figure shows the normalised distribution of delays that I modelised with an exponential distribution  f(x)=aexp(−x/b) . The  a  et  b  parameters obtained to describe each airline are given in the upper right corner of each panel. Note that the normalisation of the distribution implies that  ∫f(x)dx∼1 . Here, we do not have a strict equality since the normalisation applies the histograms but not to the model function. However, this relation entails that the  a  et  b  coefficients will be correlated with  a∝1/b  and hence, only one of these two values is necessary to describe the distributions. Finally, according to the value of either  a  or  b , it is possible to establish a ranking of the companies: the low values of  a  will correspond to airlines with a large proportion of important delays and, on the contrary, airlines that shine from their punctuality will admit hight  a  values:

```python
#Using the elbow method to find the optimum number of clusters
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 11):
    kmeans=KMeans(n_clusters=i,
    init='k-means++',
    n_init=10,
    max_iter=300,
    verbose=0,
    random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.show()
```
